```python
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('seaborn-poster')
%matplotlib inline

# declare constants
C = 4
d = d_LCDM
w = w_LCDM
B = 2*(d/(d+3*w))
A = 3


# Define parameters
f = lambda t, a: C*(A*a**(-1) + B*(1 - a**(-(d+3*w)))*a**(-1))**0.5 # ODE
h = 0.01 # Step size
t = np.arange(0, 1 + h, 2*h) # Numerical grid
a0 = 1 # Initial Condition

# Explicit Euler Method
a = np.zeros(len(t))
a[0] = a0

for i in range(0, len(t) - 1):
    a[i + 1] = a[i] + h*f(t[i], a[i])

plt.figure(figsize = (12, 8))
plt.plot(t, a, 'b--', label='Approx_LCDM')
#plt.plot(t, -np.exp(-t), 'g', label='Exact')
plt.title('Euler integration a compromise Exact Solution \
for Simple ODE')
plt.xlabel('t')
plt.ylabel('f(t)')
plt.grid()
plt.legend(loc='lower right')
plt.show()
```
